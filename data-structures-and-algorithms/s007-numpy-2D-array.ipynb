# Create 2D numpy array

```python
import numpy as np

a = [
    [11, 12, 13],
    [21, 22, 23]
]

arr =np.array(a)       # Create numpy 2D array
print("ORIGINAL ARRAY")
print(arr)

# axis=0 (row)
# axis=1 (column)

print("Dimension = {}".format(arr.ndim))
print("Shape = {}".format(arr.shape))
print("Size = {}".format(arr.size))

print("TRANSPOSE ARRAY")
transpose_arr = arr.T
print(transpose_arr)

print("RESHAPE 2X3 array to 3X2 array")
reshape_arr = arr.reshape(3,2)
print(reshape_arr)
```

# Accessing elements of 2D array

```python
print(arr[1,2])      # Approach 1 (recommended)
print(arr[1][2])     # Approach 2
```

# Slicing 2D array

```python
print(arr[0:2, 2])      # Approach 1 (recommended)
print(arr[0][0:3])     # Approach 2
```

# Statistical Functions

```python
print("SQUARE ROOT")
result = np.sqrt(arr)
print(result)
print()

print("SUM")
result = np.sum(arr)
print(result)
print()

print("MEAN")
result = np.mean(arr)
print(result)
print()

print("MIN")
result = np.min(arr)
print(result)
print()

print("MAX")
result = np.max(arr)
print(result)
print()
```

# 2D Numpy array operations

```python
x = np.array([
    [1,0],
    [0,1]
])

y = np.array([
    [2,1],
    [1,2]
])


print(x)
print(y)
```

```python
print("SCALAR ADDITION")
print(x + 1)
print()

print("SCALAR SUBTRACTION")
print(x - 1)
print()

print("SCALAR MULTIPLICATION")
print(x * 2)
print()

print("SCALAR DIVISION")
print(x / 2)
print()
```

```python
print("MATRIX ADDITION")
print(x + y)
print()
print(np.add(x,y))
print()

print("MATRIX SUBTRACTION")
print(x - y)
print()
print(np.subtract(x,y))
print()

print("MATRIX MULTIPLICATION")
print(x * y)
print()
print(np.multiply(x,y))
print()

print("MATRIX DIVISION")
print(x / y)
print()
print(np.divide(x,y))
print()
```

# Matrix (2D array) Dot Product

A 2D array is also called a **Matrix**.

**Matrix Dot Product Rule**:
1. Multiply & add elements of rows of the first matrix with the elements of the columns of the second matrix.
2. Number of columns of the first matrix **SHOULD BE EQUAL** to the number of columns of the second matrix.
3. Resulting Matrix Shape:
   * First matrix shape = **M x N** 
   * Second matrix shape = **N x P**. 
   * New Matrix Shape = **M x P**



A = 2 x 3
B = 3 x 2

AB = 2 x 2 (**Resultant Matrix**)


```python
A = np.array([
    [0,1,1], 
    [1,0,1]
])

B = np.array([
    [1,1],
    [1,1],
    [-1,1]
])

print(np.dot(A,B))
```
